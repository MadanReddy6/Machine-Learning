```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('kc_house_data.csv')


df.drop(["id","date","zipcode","lat","long","sqft_living15","sqft_lot15"], axis = 1, inplace= True)
df
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[['bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view','condition','sqft_above','sqft_basement','yr_built','yr_renovated']],df[['price','grade']])
```

```python
reg.coef_
```

```python
reg.intercept_
```
