```python
df = pd.read_csv(R"C:\Users\K Madan Mohan Reddy\Desktop\Python\ml\Almond.csv")

df.drop(["Unnamed: 0","Roundness","Eccentricity","Compactness","Extent"],axis=1,inplace= True)
df.dropna()
df["Type"].value_counts()
```

```python
pd.isnull(df["Length (major axis)"])
major_axix_mean = df["Length (major axis)"].mean()
df["Length (major axis)"].fillna(major_axix_mean,inplace=True)
nope = []
        
```

```python
df
```

```python
pd.isnull(df["Width (minor axis)"])
width_mean = df["Width (minor axis)"].mean()
df["Width (minor axis)"].fillna(width_mean,inplace= True)
width_mean
```

```python
pd.isnull(["Thickness (depth)"])
Thickness_mean = df["Thickness (depth)"].mean()
df["Thickness (depth)"].fillna(Thickness_mean, inplace= True)
df
```

```python
aspect_mean = df["Aspect Ratio"].mean()
df["Aspect Ratio"].fillna(aspect_mean, inplace=True)
df["Aspect Ratio"].info()
df
```

```python
df["Type"].info()
```

```python

#making lables frequency same 
for i , j in df.groupby("Type"):
    if len(j) < 900 :
        df.drop(j.index , axis=0, inplace=True) 
        
    elif len(j) > 900:
        x = len(j) - 900
        df.drop(j[:x].index, axis=0 , inplace= True)

```

```python

#normalization
x = df[["Length (major axis)","Width (minor axis)","Thickness (depth)","Area","Perimeter","Solidity","Aspect Ratio","Convex hull(convex area)"]]
org_mean , org_std = x.mean(), x.std()

df[["Length (major axis)","Width (minor axis)","Thickness (depth)","Area","Perimeter","Solidity","Aspect Ratio","Convex hull(convex area)"]]= (x-org_mean)/ org_std
targets = df[["Type"]]
inputs = df.drop(["Type"], axis= 1)
inputs.info()
```

```python
# one_hot_targets = pd.get_dummies(targets)

one_hot_targets = pd.get_dummies(targets)
labels = targets["Type"].unique()
labels.sort()
labels
```

```python
# converting into tensors


inputs_tf = tf.constant(inputs)
targets_tf = tf.cast(tf.constant(one_hot_targets), tf.float64)
```

```python

model = Sequential([
    layers.Dense(3),
    layers.Softmax()
])

model.build(input_shape = inputs_tf.shape)
```

```python

model.compile(
    loss = categorical_crossentropy,
    optimizer = SGD(learning_rate=0.01),
    metrics = ["accuracy"]
)
```

```python
targets_tf
```

```python
history = model.fit(
    inputs_tf,
    targets_tf,
    epochs = 1000,
    batch_size = 50
)
```

```python
history_df = pd.DataFrame(history.history)
history_df
```

```python
plt.plot(history_df["loss"],history_df["accuracy"],color = 'red')
plt.ylabel('accuracy')
plt.xlabel('loss')
plt.show()
```
