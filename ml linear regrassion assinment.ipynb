```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1
# import your data, and decide on the input and output

df = pd.read_csv('./weather.csv')[:500]
plt.scatter(df['Temperature (C)'],df['Humidity'],alpha=0.1)

df['Temperature (C)']= df['Temperature (C)'] / 10 
df.describe()

df['prediction'] = weight * df['Temperature (C)'] + bias


plt.plot(df['Temperature (C)'], df['prediction'], c='red')
plt.show()
```

```python
def mse(y,y_):
    
    return sum((y-y_)**2) /len(y)
def grad(y, y_ , x):
    return sum(-2*(y-y_)*x )/len(y),sum(-2*(y - y_))/len(y)

def pred(w , b , x):
    return w*x + b
```

```python
# step 2
# initializing the weights, biases, learning rate

weight = 0
bias = -0
```

```python
# training loop
 
feature = df['Temperature (C)']
label = df['Humidity']

lr = 0.001

for i in range(10000):
    
   
    prediction = pred(weight,bias,feature)
    loss = mse(label,prediction)
    
    print(loss)
    w_grad , b_grad = grad(label,prediction,feature)
    
    weight = weight - w_grad*lr
    bias= bias - b_grad*lr
    
    
print(loss, weight,bias)    
```
